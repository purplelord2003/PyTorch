### Matrix multiplication
1. Element-wise
2. Matrix multiplication (dot product)

For matrix multiplications:
1. The inner dimensions must match
* `(3, 2) @ (3, 2)` wont work
* `(2, 3) @ (3, 2)` will work

2. The resulting matrix has the shape of the outer dimensions
* `(2, 3) @ (3, 2) gives matrix of shape (2, 2)`

```python
torch.rand(2,3) @ torch.rand(3,2)
```

```python
# Element-wise mult
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix mult
torch.matmul(tensor, tensor)
```

### Shape errors

```python
# Shapes for matrix mult
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])
torch.mm(tensor_A, tensor_B) # torch.mm is alias for torch.matmul
```

To fix our tensor shape issues, we can manipulate shape of one of the tensors using a **transpose**

A **transpose** switches the axes or dimensions of a given tensors


```python
tensor_B, tensor_B.shape
```

```python
# Transpose
tensor_B.T, tensor_B.T.shape
```

```python
torch.matmul(tensor_A, tensor_B.T)
```

## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - return a view of an input tensor of certain shape but
keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or beside (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - adds a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted in a certain way

```python
# Creating a tensor
import torch

x = torch.arange(1.,10.)
x, x.shape
```

```python
# Reshape to add an extra dimension (number of elements must match)
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x because a view of a tensor shares same memory as the input
z[:, 0] = 5
z, x
```

```python
# Stacking tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# vstack (vertical stack)
x_vstack = torch.vstack([x,x,x,x])
x_vstack
```

```python
# hstack (horizontal stack)
x_hstack = torch.hstack([x,x,x,x])
x_hstack
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
tensor = torch.tensor([[[1,2], [3,4]]])
```

```python
tensor.shape
```

```python
tensor = torch.squeeze(tensor) # Or tensor.squeeze()
tensor
```

```python
tensor.shape
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim
tensor = torch.tensor([1,2,3,4])
tensor
```

```python
torch.unsqueeze(tensor, 0) # adding to 0th dim
```

```python
torch.unsqueeze(tensor, 0).shape
```

```python
torch.unsqueeze(tensor, 1) # adding to 1st dim
```

```python
torch.unsqueeze(tensor, 1).shape
```

```python
# torch.permute - rearranges the dimensions of the target tensor in a particular order
x = torch.rand(size=(224,224,3)) # height, width, colour channels for image tensor
x
```

```python
x.shape
```

```python
x_permuted = x.permute(2, 0, 1) # shifts axis 0->1, 1-->2, 2-->3
x_permuted.shape # channels, height, width now
```

## Reproducibility (trying to take random out of random)

In short, how a neural network learns:
`start with random numbers -> tensor operations -> update random numbers to make them better representations -> again -> again...`

To reduce randomness, concept of random seed.

Random seed flavour the randomness

```python
import torch

# Create 2 random tensors
A = torch.rand(3, 4)
B = torch.rand(3, 4)

print(A)
print(B)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set random seed
RANDOM_SEED = 42 # seed only works for one block of code
torch.manual_seed(RANDOM_SEED)
C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED) # repeat
D = torch.rand(3,4)

print(C)
print(D)
```
